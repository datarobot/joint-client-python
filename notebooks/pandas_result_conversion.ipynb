# Pandas Result Conversion
Parse a checked-in response fixture and convert it to tidy, wide, and NumPy forms.

```python
import json
from pathlib import Path

from jointfm_client import ForecastResponse

payload = json.loads(Path('tests/fixtures/forecast_quantiles_response.json').read_text())
result = ForecastResponse.from_payload(payload)
tidy = result.to_pandas_tidy()
wide = result.to_pandas_wide()
values = result.to_numpy()
tidy, wide, values.shape
```
