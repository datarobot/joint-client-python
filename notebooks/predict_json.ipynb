# Low-Level JSON Predict
Submit a checked-in V1 request payload through `JointFMClient.predict`.

```python
import json
from pathlib import Path

from jointfm_client import JointFMClient

client = JointFMClient.from_env()
payload = json.loads(Path('tests/fixtures/forecast_mean_request.json').read_text())
payload['model_version'] = client.health(cache=True).model_version
response = client.predict(payload)

print('Payload:')
print(json.dumps(payload, indent=2))
print('\nResponse:')
print(json.dumps(response, indent=2))
```
